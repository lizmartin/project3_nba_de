Fantasy score - Julia

```python
# Creating dataframe for fantasy score section
fantasy_score = games_details[['PLAYER_NAME',
 'REB',
 'AST',
 'STL',
 'BLK',
 'TO',
 'PTS']]
fantasy_score
```

```python
# Grouping by player so we can get the total amount for each stat
fantasy_score_grouped = fantasy_score.groupby('PLAYER_NAME').sum().reset_index()
fantasy_score_grouped
```

```python
# Creating dataframe with average mins to merge with fantasy_score_grouped
avg_mins = games_details[['PLAYER_NAME','MIN']].groupby('PLAYER_NAME').mean()
avg_mins
```

```python
fantasy_merged = pd.merge(fantasy_score_grouped,avg_mins, on='PLAYER_NAME',how='inner')
fantasy_merged
```

Fantasy points breakdown:
PTS = 1 Points
REB = 1.2 Points
AST = 1.5 Points
STL = 3 Points
BLK = 3 Points
TO = -1 Points

```python
# Function to perform calculations on each column based on fantasy point system
def fantasy_score_calc(df):
    df['REB'] = df['REB'] * 1.2
    df['AST'] = df['AST'] * 1.5
    df['STL'] = df['STL'] * 3
    df['BLK'] = df['BLK'] * 3
    df['TO'] = df['TO'] * -1
    df['PTS'] = df['PTS'] * 1
    
```

```python
# Run df through function
fantasy_score_calc(fantasy_score_grouped)
```

```python
fantasy_score_grouped
```

```python
# Function to perform total points
def total_pts_calc(df):
    df['Fantasy_Score'] = df[
    'REB'] + df['AST'] + df[
        'AST'] + df['STL'] + df[
            'BLK'] + df['TO'] + df['PTS']
```

```python
total_pts_calc(fantasy_score_grouped)
```

```python
# Find players with highest fantasy score
best_fantasy_players = fantasy_score_grouped.sort_values(
    by='Fantasy_Score', ascending=False)
best_fantasy_players.head(12)
```

```python
# Addings mins as well to last dataframe
best_fantasy_players = pd.merge(best_fantasy_players,avg_mins,on='PLAYER_NAME',how='inner')
best_fantasy_players
```
