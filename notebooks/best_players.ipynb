```python
# Dependencies
import pandas as pd
from pathlib import Path

# CSV paths
games_details_path = Path('../data/cleaned_data/games_details_cleaned.csv')
games_path = Path('../data/cleaned_data/games_cleaned.csv')
teams_path = Path('../data/cleaned_data/teams_cleaned.csv')
players_path = Path('../data/cleaned_data/players_cleaned.csv')
ranking_path = Path('../data/cleaned_data/ranking_cleaned.csv')

# Creating dataframes
games_details = pd.read_csv(games_details_path)
games = pd.read_csv(games_path)
teams = pd.read_csv(teams_path)
players = pd.read_csv(players_path)
ranking = pd.read_csv(ranking_path)
```

```python
# Getting rid of players who aren't starters
starters = games_details.dropna(subset=['START_POSITION'])
starters
```

```python
# Create a dataframe with the columns relevant to fantasy points
fantasy_score = starters[['PLAYER_NAME',
 'REB',
 'AST',
 'STL',
 'BLK',
 'TO',
 'PTS']]
fantasy_score
```

```python
# Grouping by player so we can get the total amount for each stat
fantasy_score_grouped = fantasy_score.groupby('PLAYER_NAME').sum().reset_index()
fantasy_score_grouped
```

PTS = 1 Points
REB = 1.2 Points
AST = 1.5 Points
STL = 3 Points
BLK = 3 Points
TO = -1 Points

```python
# Perform calculations on each column based on fantasy point system
fantasy_score_grouped['REB'] = fantasy_score_grouped['REB'] * 1.2
fantasy_score_grouped['AST'] = fantasy_score_grouped['AST'] * 1.5
fantasy_score_grouped['STL'] = fantasy_score_grouped['STL'] * 3
fantasy_score_grouped['BLK'] = fantasy_score_grouped['BLK'] * 3
fantasy_score_grouped['TO'] = fantasy_score_grouped['TO'] * -1
fantasy_score_grouped['PTS'] = fantasy_score_grouped['PTS'] * 1
fantasy_score_grouped
```

```python
# Create a new column with the total fantasy points
fantasy_score_grouped['Fantasy_Score'] = fantasy_score_grouped[
    'REB'] + fantasy_score_grouped['AST'] + fantasy_score_grouped[
        'AST'] + fantasy_score_grouped['STL'] + fantasy_score_grouped[
            'BLK'] + fantasy_score_grouped['TO'] + fantasy_score_grouped['PTS']

fantasy_score_grouped
```

```python
# Find players with highest fantasy score
best_fantasy_players = fantasy_score_grouped.sort_values(
    by='Fantasy_Score', ascending=False)
best_fantasy_players.head(12)
```
